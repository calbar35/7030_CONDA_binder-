# Importing the dataset

```r
dataset = read.csv('regrex1.csv')
```

# Fitting Linear Regression to the dataset

```r
model = lm(formula = y ~ x,
               data = dataset)
               
```

# Summary analysis of the model 

```r
summary(model)
```

# Visualizing the Linear Regression results

```r
library(ggplot2)
ggplot() +
  geom_point(aes(x = dataset$y, y = dataset$x),
             colour = 'red') +
  geom_line(aes(x = dataset$y, y = predict(model, newdata = dataset)),
            colour = 'blue') +
  ggtitle('x vs y') +
  xlab('y') +
  ylab('x')
```
